## Cohen's Kappa

```python
folder = "../../user_study/TSE/"

fpath_1 = folder + "gender-labelled-J.csv"
fpath_2 = folder + "gender-labelled-SSQ.csv"
annotation_1 = pd.read_csv(fpath_1)
annotation_2 = pd.read_csv(fpath_2)
```

```python
df1 = annotation_1
df2 = annotation_2
```

### Preprocessing


```python
df1.groupby("sentiment").count()
```

```python
df1["sentiment"] = df1["sentiment"].replace("Not sure", "Neutral") 
df1.groupby("sentiment").count()
```

```python
df2.groupby("sentiment").count()
```

```python
df2["sentiment"] = df2["sentiment"].replace("neutral", "Neutral") 
df2.groupby("sentiment").count()
```

```python
df2[df2["sentiment"] == "Neutral"].head()
```

```python
df2[df2["is_make_sense"] == "No"]
```

```python
# categories
cat = {"Positive", "Negative", "Neutral"}
# matrix
mat = {}

for c1 in cat :
    mat[c1] = {}
    for c2 in cat :
        mat[c1][c2] = 0

for i in range(len(df1)) :
    for c1 in cat :
        for c2 in cat :
            if df1["sentiment"][i] == c1 and df2["sentiment"][i] == c2 :
                mat[c1][c2] += 1
```

```python
mat
```

```python
total = 0
agreement = 0
for c1 in cat :
    for c2 in cat :
        total += mat[c1][c2]
        if c1 == c2 :
            agreement += mat[c1][c2]

print("Total: ", total)
print("Agreement: ", agreement)
```

```python
ef1 = {}
for c1 in cat :
    _ef1 = 0
    for c2 in cat :
        _ef1 += mat[c1][c2]
    ef1[c1] = _ef1

ef1    
```

```python
ef2 = {}
for c1 in cat :
    _ef2 = 0
    for c2 in cat :
        _ef2 += mat[c2][c1]
    ef2[c1] = _ef2
ef2
```

```python
ef = 0
for c in cat :
    ef += ef1[c] * ef2[c] / total
ef
```

```python
ck = (agreement - ef)/(total - ef)
print("Cohen's Kappa: ",  ck)
```

# TPR

```python
df1["mod"] = [x % 3 for x in list(df1.index)]
df2["mod"] = [x % 3 for x in list(df2.index)]
```

### Separate data into male and female

```python
dff1 = df1[df1["mod"] == 1].reset_index()
dfm1 = df1[df1["mod"] == 2].reset_index()

dff1
```

```python
dff2 = df2[df2["mod"] == 1].reset_index()
dfm2 = df2[df2["mod"] == 2].reset_index()

dff2
```

### TPR from 1st labeller

```python
not_make_sense = 0
coherent = 0

for i in range(len(dff1)) :
    if dff1["is_make_sense"][i] == "Yes" and dfm1["is_make_sense"][i] == "Yes":
        if dff1["sentiment"][i] == dfm1["sentiment"][i] :
            coherent += 1
    else :
        not_make_sense += 1
```

```python
print("TPR 1st labeller: ", coherent/len(dff1))
print("Not make sense: ", not_make_sense)
```

### TPR from 2nd labeller

```python
not_make_sense = 0
coherent = 0

for i in range(len(dff2)) :
    if dff2["is_make_sense"][i] == "Yes" and dfm2["is_make_sense"][i] == "Yes":
        if dff2["sentiment"][i] == dfm2["sentiment"][i] :
            coherent += 1
    else :
        not_make_sense += 1
```

```python
print("TPR 2nd labeller: ", coherent/len(dff2))
print("Not make sense: ", not_make_sense)
```
