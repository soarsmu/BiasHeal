### Data Preparation for RQ1 Exp 1 Debiasing BiasFinder

```python
import pandas as pd
import numpy as np
import math
import os
import random 
```

```python
data_dir = "gc_imdb"
df = pd.read_csv("../data/" + data_dir + "/test.csv", header=None, sep="\t", names=["label", "mutant", "template", "original", "identifier", "type", "gender", "country"])
df = df.sample(frac=1, random_state=123).reset_index(drop=True)
```

### Unbalanced Male-Female

```python
mid = int(len(df)/2)
train = df[:-mid]
test = df[-mid:]
```

```python
train.to_csv("../data/gc_imdb/50persen/train.csv", index=None, header=None, sep="\t")
test.to_csv("../data/gc_imdb/50persen/test.csv", index=None, header=None,sep="\t")
```

### Balanced Male-Female

```python
mdf = df[df["gender"] == "male"]
fdf = df[df["gender"] == "female"]
```

```python
mdf = mdf.sample(frac=1, random_state=123).reset_index(drop=True)
fdf = fdf.sample(frac=1, random_state=123).reset_index(drop=True)
```

```python
mid = int(len(mdf)/2)
train = pd.concat([mdf[:-mid], fdf[:-mid]])
test = pd.concat([mdf[-mid:], fdf[-mid:]])
```

```python
train.to_csv("../data/gc_imdb/50persenbalanced/train.csv", index=None, header=None, sep="\t")
test.to_csv("../data/gc_imdb/50persenbalanced/test.csv", index=None, header=None,sep="\t")
```
