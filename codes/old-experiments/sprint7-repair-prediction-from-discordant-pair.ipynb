# Number of Discordant Pairs Country

Discordant pair is a pair contain of male-female and its prediction, such that the Sentiment Analysis produce a different prediction. 
Example of discordant pair from country: 

`<(name from Indonesia, prediction), (name from USA, prediction)>`

`<(“Paijo is angry”, 1), (“John is angry”, 0)>`

```python
import pandas as pd
import numpy as np
import math
import time
```

```python
eval_dir = "gc_imdb"
# df = pd.read_csv("../data/" + eval_dir + "/test.csv", header=None, sep="\t", names=[0, 1, "mutant", "template", "gender", "label", "country"])
# df = pd.read_csv("../data/" + eval_dir + "/test.csv", header=None, sep="\t", names=[0, 1, "mutant", "template", "occupation"])
df = pd.read_csv("../data/" + eval_dir + "/test.csv", header=None, sep="\t", names=["label", "mutant", "gender", "template", "template_id"])
df
```

```python
def read_txt(fpath):
    pred = []
    file = open(fpath)
    lines = file.readlines()
    for l in lines :
        pred.append(int(l))
    file.close()
    
    return pred
```

```python
output_dir = "gc_imdb"

result_dir = "../result/" + output_dir + "/"

path = result_dir + "results_data.txt"

pred = read_txt(path)

print(len(pred))
```

```python
df["prediction"] = pred
```

```python
# df = df.drop(columns=[0,1])
df.head()
```

## Separate the Original Text with Mutant Text

```python
dft = df[df["gender"] == "template"]
df = df[df["gender"] != "template"]

# dft = df[df.index % 311 == 0]
# df = df[df.index % 311 != 0]

dft = dft.reset_index(drop=True)
df = df.reset_index(drop=True)
```

## Find Majority Prediction from each Template

```python
dft
```

```python
df
```

```python
max_size = max(df.groupby("template_id").sum()["label"])
max_size
```

```python
prediction_count = df.groupby("template_id").sum()["prediction"]
prediction_count
```

```python
majority_prediction = []
for p in prediction_count:
    if p * 2 < max_size :
        majority_prediction.append(0)
    else :
        majority_prediction.append(1)
```

```python
dft = dft.sort_values(by=["template_id"])
dft["majority_prediction"] = majority_prediction
dft.drop(columns=["mutant", "gender"])
```

```python
mp = dict(zip(dft["template_id"].values, dft["majority_prediction"].values))

def get_majority_prediction_from_template_id(id):
    return mp[id]
```

#### Performance for "label" vs "prediction"

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

y_test = dft["label"]
y_pred = dft["prediction"]
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)

print(cm)
```

#### Performance for "label" vs "majority_prediction"

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

y_test = dft["label"]
y_pred = dft["majority_prediction"]
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## Create Discordant Pairs based on Gender

```python
gb = df.groupby("template_id")
```

```python
gb.count()
```

```python
len(gb.size())
```

```python
start = time.time()

identifier = "gender"

mutant_text_1 = []
mutant_text_2 = []
prediction_1 = []
prediction_2 = []
identifier_1 = []
identifier_2 = []
template = []
template_id = []
label = []
for i in range(len(gb.size())) :
# for i in range(2000, 3000) :
    data = gb.get_group(i)
    dc = data.groupby(identifier)
    for k1, v1 in dict(iter(dc)).items() :
        for k2, v2 in dict(iter(dc)).items() :
            if k1 != k2 :
                for m_1, p_1, i_1, t, t_id, l in zip(v1["mutant"].values, v1["prediction"].values, v1[identifier].values, v1["template"].values, v1["template_id"].values, v1["label"].values) :
                    for m_2, p_2, i_2 in zip(v2["mutant"].values, v2["prediction"].values, v2[identifier].values) :
                        mutant_text_1.append(m_1)
                        prediction_1.append(p_1)
                        identifier_1.append(i_1)
                        mutant_text_2.append(m_2)
                        prediction_2.append(p_2)
                        identifier_2.append(i_2)
                        template.append(t)
                        template_id.append(t_id)
                        label.append(l)

end = time.time()
print("Execution time: ", end-start)
```

```python
df = pd.DataFrame(data={"mutant_1" : mutant_text_1, 
                        "mutant_2" : mutant_text_2, 
                        "prediction_1": prediction_1, 
                        "prediction_2" : prediction_2, 
                        "identifier_1": identifier_1, 
                        "identifier_2" : identifier_2, 
                        "template": template,
                        "template_id": template_id,
                        "label" : label
                       })

df.head()
```

```python
df["discordant"] = df["prediction_1"] != df["prediction_2"]
df
```

```python
print("Number of Discordant Pairs: ", len(df[df["discordant"] == True]))
```

```python
print("Discordant Pairs Rate: ", len(df[df["discordant"] == True]) / len(df) )
```

```python
d = df[df["discordant"] == True]
d = d.drop(columns=["discordant"])
d
```

#### Add Majority Prediction

```python
d["majority_prediction"] = d["template_id"].apply(get_majority_prediction_from_template_id)
d
```

#### How good is the repair?

```python
n_repair = len(d)
print("Number of Repairment for Discordant Pairs: ", n_repair)
```

```python
d["good_repair"] = d["label"] == d["majority_prediction"]
d.head()
```

```python
n_correct_repair = len(d[d["good_repair"] == True])
print("Number of Correct Repairment for Discordant Pairs: ", n_correct_repair)
```

#### Success Rate

```python
print("Success rate: ", n_correct_repair/n_repair)
```
