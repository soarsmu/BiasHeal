```python
import pandas as pd
import numpy as np
import math
import time
```

```python
data_dir = "../data/rq2/biasfinder/btc/"

df = pd.read_csv(data_dir + "full.csv")
df
```

```python
df = df.iloc[:,[0,1,6,7]]
df
```

```python
def read_txt(fpath):
    pred = []
    file = open(fpath)
    lines = file.readlines()
    for l in lines :
        pred.append(int(l))
    file.close()
    
    return pred
```

```python
output_dir = "rq2/biasfinder"

result_dir = "../result/" + output_dir + "/"

path = result_dir + "results_data.txt"

pred = read_txt(path)

print(len(pred))
```

```python
df["prediction_1"] = pred[:len(df)]
df["prediction_2"] = pred[len(df):]
df
```

```python
df
```

```python
df["btc"] = df["prediction_1"] != df["prediction_2"]
df
```
