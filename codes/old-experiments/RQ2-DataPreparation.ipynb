```python
dirname = "../data/gc_imdb/"
# dirname = "../data/imdb_using_eec_name/"
df = pd.read_csv(dirname + "test.csv", sep="\t", names=["label", "mutant", "template", "original", "identifier", "type", "gender", "country"])
```

### IMDB Original Text

```python
# take original text
do = df.iloc[:,[0, 3]]
do = do.drop_duplicates().reset_index(drop=True)
do
```

```python
import os 

data_dir = "../data/rq2/imdb_original/"

if not os.path.exists(data_dir) :
    os.makedirs(data_dir)

do.to_csv(data_dir + "train.csv", index=None, header=None, sep="\t")
```

### IMDB Original Text + Mutant 20% with Train 80%

```python
df
```

```python
df.groupby("gender").get_group("male").sample(frac=1, random_state=123)
```

```python
portion = 0.2
seed = 123
size = int(portion * len(df.groupby("gender").get_group("male")))
mtrain = df.groupby("gender").get_group("male").sample(frac=1, random_state=seed).reset_index(drop=True)[:size]
mtest = df.groupby("gender").get_group("male").sample(frac=1, random_state=seed).reset_index(drop=True)[size:]

ftrain = df.groupby("gender").get_group("female").sample(frac=1, random_state=seed).reset_index(drop=True)[:size]
ftest = df.groupby("gender").get_group("female").sample(frac=1, random_state=seed).reset_index(drop=True)[size:]

train = pd.concat([do, mtrain, ftrain])
train = train.sample(frac=1, random_state=seed)
test = pd.concat([mtest, ftest])
```

```python
data_dir = "../data/rq2/imdb_original_mutant_20/"

if not os.path.exists(data_dir) :
    os.makedirs(data_dir)

train.to_csv(data_dir + "train.csv", index=None, header=None, sep="\t")
```

```python
data_dir = "../data/rq2/mutant_80/"

if not os.path.exists(data_dir) :
    os.makedirs(data_dir)

test.to_csv(data_dir + "test.csv", index=None, header=None, sep="\t")
```
