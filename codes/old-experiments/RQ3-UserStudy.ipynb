```python
data_dir = "../data/rq2/biasfinder/btc/"

df = pd.read_csv(data_dir + "full.csv")
df
```

```python
df = df.sample(frac=1, random_state=123)
df
```

```python
sample_size = 384
sample = df.iloc[:sample_size].reset_index(drop=True)
```

```python
sample
```

```python
is_make_sense = [None] * 2 * len(sample)
sentiment = [None] * 2 * len(sample)
comment = [None] * 2 * len(sample)
text = [None] * 2 * len(sample)
i = 0
for index, row in sample.iterrows() :
    text[i*2] = sample.iloc[i]["mutant_1"]
    text[i*2+1] = sample.iloc[i]["mutant_2"]
    i += 1
```

```python
df = pd.DataFrame(data={"text" : text, "sentiment": sentiment, "is_make_sense" : is_make_sense, "comment" : comment})
df
```
